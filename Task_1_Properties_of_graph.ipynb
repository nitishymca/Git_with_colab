**Basic Properties of Graph**

```python
import networkx as nx
```

```python
# Create a sample graph (you can replace this with your own graph)
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (4, 5)])

# Calculate node degrees
node_degrees = dict(G.degree())
print("Node degrees:", node_degrees)
```

```python
# Calculate common neighbors between two nodes
node1 = 1
node2 = 2
common_neighbors = list(nx.common_neighbors(G, node1, node2))
print(f"Common neighbors between nodes {node1} and {node2}:", common_neighbors)

```

```python
# Calculate centrality measures (e.g., degree centrality, betweenness centrality)
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)

print("Degree centrality:", degree_centrality)
print("Betweenness centrality:", betweenness_centrality)
```

**Graph property: Degree distribution, No. of Edges/Nodes, Visualize**

```python
# Graph

G = nx.Graph()

edges = [(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)]
G.add_edges_from(edges)

# Calculate and print graph properties
num_nodes = G.number_of_nodes()
print(f"Number of nodes: {num_nodes}")
print("----------------------------------")

num_edges = G.number_of_edges()
print(f"Number of edges: {num_edges}")
print("----------------------------------")

degree_distribution = [degree for node, degree in G.degree()]
print(f"Degree distribution: {degree_distribution}")
```

```python
# pos = nx.spring_layout(G)
# pos = nx.circular_layout(G)
# pos = nx.kamada_kawai_layout(G)
# pos = nx.random_layout(G)
pos =  nx.shell_layout(G)

nx.draw(G, pos, with_labels=True, node_size=300, node_color='skyblue', font_size=10, font_color='black')

plt.title("----- Graph Visualization ------")

plt.show()
```
